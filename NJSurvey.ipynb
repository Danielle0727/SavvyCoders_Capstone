```python
#If using only specific columns: usecols =[''] in pd.read_csv after file name

#print all column names
#for col in df.columns:
    #print(col)

#Import CSV File
import pandas as pd

df = pd.read_csv(
    'C:\\Users\Danielle\Documents\DAP\Data Science and Capstone\MASTERChildlessnessNJ.csv')

#---------------------------------------------

#Sort by age
df.sort_values(by = ['Age'], inplace = True)
#print(df)

#Finding distinct info in columns
Genders = df['Gender'].unique()
Ages = df['Age'].unique()
Employment = df['Employment Status'].unique()
Race = df['Race/Ethnicity'].unique()
Relationship = df['Relationship Status'].unique()
# print(Genders)
# print(Ages)
# print(Employment)
# print(Race)
# print(Relationship)

#Replace blank data in Q7 with "Blank"
df['Q7'].fillna('Blank', inplace=True)
#print(df['Q7'].to_string())

#Drop Nulls in original df
df.dropna(inplace=True)
#print(df)

#Q11: 0.0 values. Verify they weren't dropped.
print(df['Q11'].to_string(index=False))

#Replace Likert Scale numerical values w/actual meaning:
    '''
    0 with 'Zero'
    1 with 'Strongly Disagree'
    2 with 'Disagree'
    3 with 'Neutral'
    4 with 'Agree'
    5 with 'Strongly Disagree'
    *Some of numerical values were identified as integers
    and others as floats.
    '''

df.replace(to_replace=0, value='Zero', inplace=True)
df.replace(to_replace=1, value='Strongly Disagree', inplace=True)
df.replace(to_replace=2, value='Disagree', inplace=True)
df.replace(to_replace=3, value='Neutral', inplace=True)
df.replace(to_replace=4, value='Agree', inplace=True)
df.replace(to_replace=5, value='Strongly Agree', inplace=True)
#print(df)
#print(df['Q11'])

#Print New Changes:
print(df)

#Save data file to csv
df.to_csv('NJSurvey.csv')
```
